```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
house=pd.read_csv("./../Stars.csv")
```

```python
house.head()
```

```python
plt.title('Vertical')
plt.bar(house["Spectral_Class"],house["Temperature"])
plt.show()
```

```python
plt.title('Gorizontal')
plt.barh(house["Color"],house["Temperature"])
plt.show()
```

```python
plt.title('Mnogoryadnie')
bw=0.3
plt.bar(house["Type"],house["R"],color='r')
plt.bar(house["Type"]+bw,house["Temperature"],bw,color='b')
plt.bar(house["Type"]+2*bw,house["L"],bw,color='g')
plt.show()
```

```python
plt.title('Slozhennie')
plt.bar(house["Type"],house["Temperature"],color='r')
plt.bar(house["Type"],house["L"],color='b',bottom=house["Temperature"])
plt.bar(house["Type"],house["R"],color='g',bottom=(house["L"]+house["Temperature"]))
plt.show()
```
